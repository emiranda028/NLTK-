## 1.- Import Librerias

```python
import re
import spacy as sp
import nltk
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from urllib.request import urlopen
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer("[\w]+")
import string  
stopwords = stopwords.words('spanish')
import seaborn as sns
sns.set()
%config InlineBackend.figure_format = 'retina'
plt.style.use("seaborn-white")
```

## 2.- Carga de Datos

```python
df_macri= pd.read_fwf('Discursos-de-MM-2016(20200710_103752).txt')
df_macri =  pd.DataFrame(df_macri["Discurso"])
df_macri["owner"] = 1

df_cfk = pd.read_fwf('Discursos-de-CFK-2012-2015(20200710_103745).txt')
df_cfk = pd.DataFrame(df_cfk["Discurso"])
df_cfk["owner"] = 0

df = pd.concat([df_cfk, df_macri])

df.head(5)
#Validar que las columnas borradas se puedan incluir en una sola columna
```

## 4.- Vectorizar

#### Normalizar

```python
##Lemmatizador
def lemmatizer(text):
    nlp = sp.load('es_core_news_sm')
    doc = nlp(text)
    return ' '.join([word.lemma_ for word in doc])
```

```python
titular_list=[]
for titular in df["Discurso"]:
        
    titular = re.sub("[^A-Za-zÁ-Úá-ú \t]","",str(titular))

     # Tokenizamos para separar las palabras
    titular=tokenizer.tokenize(titular)
    
#     # Aplicamos el Lemmatizer (Esto puede tardar un ratito)
#     titular = [lemmatizer(x) for x in titular]
    
   #Sacar signos 
    titular = [char for char in titular if char not in string.punctuation]
    
    # Eliminamos las palabras d emenos de 3 letras
    titular = [palabra for palabra in titular if len(palabra)>2]
    # Sacamos las Stopwords
    titular = [palabra for palabra in titular if  palabra not in stopwords]

    # Por ultimo volvemos a unir el titular
    titular=" ".join(titular)
    #dataset["titular_normalizado"] = titular_list
    titular_list.append(titular)
```

```python
df["discurso_lema"] = titular_list 
df.tail()
```

#### Aplicar Frecuencia de palabras 

#### Aplicar tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
ngram_range = (1,2)
max_features = 2000

vectorizer = TfidfVectorizer(stop_words=stopwords,
                             max_features= max_features,
                             ngram_range=ngram_range,
                            )
tfidf_matrix = vectorizer.fit_transform(df["discurso_lema"]) 
print(tfidf_matrix.shape)
print(tfidf_matrix.todense()[0])
print(vectorizer.vocabulary_)
```

## 5.- Iniciar Modelo

```python
from sklearn.model_selection import train_test_split
X = tfidf_matrix.toarray()
y = df['owner'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

def confusion(ytest,y_pred):
    names=["CFK","MM"]
    cm=confusion_matrix(ytest,y_pred)
    f,ax=plt.subplots(figsize=(8,8))
    sns.heatmap(cm,annot=True,linewidth=.5,linecolor="r",fmt=".0f",ax=ax)
    plt.xlabel("y_pred")
    plt.ylabel("y_true")
    ax.set_xticklabels(names)
    ax.set_yticklabels(names)
    plt.show()

    return
```

#####  5.1-SVM

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(C = 1)
svc.fit(X_train,y_train)
svc.score(X_test,y_test)
```

```python
y_pred=svc.predict(X_test)
confusion(y_test,y_pred)
```

##### 5.2-Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 50, random_state=42)
rf.fit(X_train,y_train)
print("acc: ",rf.score(X_test,y_test))
```

```python
y_pred=rf.predict(X_test)
confusion(y_test,y_pred)
```

```python
# Le preguntamos la importancia de cada feature (cada palabra)
importances = rf.feature_importances_
# Tomamos la lista de palabras
all_words = vectorizer.get_feature_names()
columns = all_words

# Ordenamos por importnacia y tomamos las 20 primeras
indices = np.argsort(importances)[::-1]
indices = indices[:10]
selected_columns = [columns[i] for i in indices]
selected_importances = importances[indices]

# Por ultimo graficamos
plt.figure(figsize = (20,10))
sns.barplot(selected_columns, selected_importances)
plt.show()
```
